```python
import cv2
import numpy as np
import random

def zoom_image(img, zoom_range=(0.10, 0.15), zoom_in=True):
    h, w = img.shape[:2]

    # Pick a random zoom factor between 10% and 15%
    zoom_factor = 1 + random.uniform(*zoom_range) if zoom_in else 1 - random.uniform(*zoom_range)

    # Resize the image
    new_h, new_w = int(h * zoom_factor), int(w * zoom_factor)
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)

    if zoom_in:
        # Crop the center to original size
        start_x = (new_w - w) // 2
        start_y = (new_h - h) // 2
        zoomed = resized[start_y:start_y+h, start_x:start_x+w]
    else:
        # Pad the image to original size
        pad_x = (w - new_w) // 2
        pad_y = (h - new_h) // 2
        zoomed = cv2.copyMakeBorder(resized, pad_y, h-new_h-pad_y, pad_x, w-new_w-pad_x,
                                    borderType=cv2.BORDER_CONSTANT, value=[0,0,0])

    return zoomed

# ---- Example usage ----
img = cv2.imread("input.jpg")

# Zoom in
zoomed_in = zoom_image(img, zoom_range=(0.10, 0.15), zoom_in=True)
cv2.imwrite("zoomed_in.jpg", zoomed_in)

# Zoom out
zoomed_out = zoom_image(img, zoom_range=(0.10, 0.15), zoom_in=False)
cv2.imwrite("zoomed_out.jpg", zoomed_out)
```

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def correct_scale_with_ssim(ground_truth, zoomed, scale_range=(0.85, 1.15), step=0.01):
    h, w = ground_truth.shape[:2]
    best_score = -1
    best_img = None
    best_scale = 1.0

    # Try different scales
    for scale in np.arange(scale_range[0], scale_range[1] + step, step):
        new_w, new_h = int(zoomed.shape[1] * scale), int(zoomed.shape[0] * scale)
        resized = cv2.resize(zoomed, (new_w, new_h), interpolation=cv2.INTER_LINEAR)

        # Center-crop or pad to match ground_truth size
        if resized.shape[0] >= h and resized.shape[1] >= w:
            # Crop
            start_y = (resized.shape[0] - h) // 2
            start_x = (resized.shape[1] - w) // 2
            resized = resized[start_y:start_y+h, start_x:start_x+w]
        else:
            # Pad
            top = (h - resized.shape[0]) // 2
            bottom = h - resized.shape[0] - top
            left = (w - resized.shape[1]) // 2
            right = w - resized.shape[1] - left
            resized = cv2.copyMakeBorder(resized, top, bottom, left, right,
                                         borderType=cv2.BORDER_CONSTANT, value=[0,0,0])

        # Compute SSIM (grayscale for robustness)
        gray_gt = cv2.cvtColor(ground_truth, cv2.COLOR_BGR2GRAY)
        gray_resized = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
        score = ssim(gray_gt, gray_resized)

        if score > best_score:
            best_score = score
            best_img = resized
            best_scale = scale

    return best_img, best_scale, best_score

# ---- Example Usage ----
gt = cv2.imread("your path")
zoomed = cv2.imread("your path")

corrected, scale, score = correct_scale_with_ssim(gt, zoomed)

print(f"Best scale factor: {scale:.3f}, SSIM: {score:.4f}")
cv2.imwrite("corrected.jpg", corrected)
```

```python
import cv2
import os
import numpy as np
from skimage.metrics import structural_similarity as ssim

# ---------- SSIM scale correction function ----------
def correct_scale_with_ssim(ground_truth, zoomed, scale_range=(0.85, 1.15), step=0.01):
    h, w = ground_truth.shape[:2]
    best_score = -1
    best_img = None
    best_scale = 1.0

    for scale in np.arange(scale_range[0], scale_range[1] + step, step):
        new_w, new_h = int(zoomed.shape[1] * scale), int(zoomed.shape[0] * scale)
        resized = cv2.resize(zoomed, (new_w, new_h), interpolation=cv2.INTER_LINEAR)

        # Match size with ground_truth by cropping or padding
        if resized.shape[0] >= h and resized.shape[1] >= w:
            start_y = (resized.shape[0] - h) // 2
            start_x = (resized.shape[1] - w) // 2
            resized = resized[start_y:start_y+h, start_x:start_x+w]
        else:
            top = (h - resized.shape[0]) // 2
            bottom = h - resized.shape[0] - top
            left = (w - resized.shape[1]) // 2
            right = w - resized.shape[1] - left
            resized = cv2.copyMakeBorder(resized, top, bottom, left, right,
                                         borderType=cv2.BORDER_CONSTANT, value=[0,0,0])

        # Compute SSIM
        gray_gt = cv2.cvtColor(ground_truth, cv2.COLOR_BGR2GRAY)
        gray_resized = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
        score = ssim(gray_gt, gray_resized)

        if score > best_score:
            best_score = score
            best_img = resized
            best_scale = scale

    return best_img, best_scale, best_score


# ---------- Main batch process ----------
ref_folder = "Reference"
zoom_folder = "Zoom in"
output_folder = "Zoom corrected"

os.makedirs(output_folder, exist_ok=True)

# Loop over reference images
for fname in os.listdir(ref_folder):
    ref_path = os.path.join(ref_folder, fname)
    zoom_path = os.path.join(zoom_folder, fname)

    if not os.path.exists(zoom_path):
        print(f"Zoomed version missing for {fname}, skipping...")
        continue

    ref_img = cv2.imread(ref_path)
    zoom_img = cv2.imread(zoom_path)

    corrected, scale, score = correct_scale_with_ssim(ref_img, zoom_img)

    out_path = os.path.join(output_folder, fname)
    cv2.imwrite(out_path, corrected)

    print(f"Saved corrected {fname} | scale={scale:.3f} | SSIM={score:.4f}")
```
